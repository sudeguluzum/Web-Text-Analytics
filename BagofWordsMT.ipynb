**Metin Temsili**, bir metni sayısal veya başka türde bir formatta temsil etme işlemidir


*   bilgisayarın anlayabilmesi
*   öznitelik çıkarma
*   model eğitimi


**Metin Temsili Yöntemleri**
1.   bag of words BoW
2.   td idf
3.   ngram
4.   word embeddings
5.   transformers tabanlı metin temsili


**Bag of Words BoW**
Her kelime bir "özellik" (feature) olarak alınır ve kelime frekansları sayılır.
Kelimelerin geçtiği ve kaç kez geçtiği önemlidir.


*   Metni tokenlera (kelimelere) ayrılır
*   Metindeki tüm benzersiz kelimelerden bir sözlük olulturur
*   Her metin için, bu sözlükteki kelimelerin frekansı sayılır


```python
#count vectorizer içeriye aktar
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#veri seti oluşturma
documents=["we love data science", "data science is fun"]
```

```python
#vectorizer tanımla
vectorizer= CountVectorizer()
```

```python
#metni sayısal vektörlere çevir
X= vectorizer.fit_transform(documents)
```

```python
# kelime kümesi oluşturma
feature_names= vectorizer.get_feature_names_out()
print("Kelime kümesi:", feature_names)
```

```python
#vektör temsili
vektor_temsili= X.toarray()

print("Vektör temsili:", vektor_temsili)
```
