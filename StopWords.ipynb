```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords") #farklı dillerde en çok kullanılan stop words içeren veri seti
```

```python
# inglizce stop words analizi (nltk)
stop_words_eng= set(stopwords.words("english"))

# örnek ingilizce metin
text= "this is an example of removing stop words from a short sentence."
text_list= text.split() #kelime kelime ayırma

#eger word ingilizce stop words listesinde (stop_words_eng) yoksa, bu kelimeyi listeye ekliyoruz
without_stopwords= [word for word in text_list if word.lower() not in stop_words_eng]
print("Stopwords olmadan text:", without_stopwords)
```

```python
# Türkçe stop words analizi (nltk)
stop_words_tr= set(stopwords.words("turkish"))

#örnek Türkçe metin
metin="O bu sabah okula gitti ve arkadaşları ile birlikte ders çalıştı."
metin_list= metin.split()

without_stopwords_tr= [word for word in metin_list if word.lower() not in stop_words_tr]
print("Stopword'süz metin:", without_stopwords_tr)
```

```python
#Kütüphanesiz stop words çıkarımı
#stop words listesi oluştur
tr_stopwords=["için","bu","ile","mu","mi","özel"]

#örnek Tr metin
metin="Bu uygulamada amacımız bu metinde bulunan özel karekterler ile durdurma kelimelerini elemek"

on_isleme=[word for word in metin.split() if word.lower() not in tr_stopwords]

print("Stopwords olmadan metin:", on_isleme)
```
