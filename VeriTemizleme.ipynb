```python
import re
import string
from bs4 import BeautifulSoup
```

```python
ornek_metin="""
<div> Bugun havva cok guzel !!!
Ziyaret et: https://wwwwhava.com :)
Bu #yaz $100'a mal oldu...
</div>
"""
print (ornek_metin)
```

```python
#html etiketlerini kaldırmak
metin = BeautifulSoup(ornek_metin, "html.parser").get_text()
print("Temizlenmiş Metin: ",metin)
```

```python
# URL'leri kalırmak için yapılacak işlemler
metin = re.sub(r"http\S+", "",metin,flags=re.MULTILINE)
print("Temizlenmiş metin:",metin)
```

```python
# metinden yer gereksiz boşlukları kaldırmak
metin = " ".join(metin.split()) # "" yazdığımızda hepsini birleşik yazıyor o yüzden " " içinde boşluk bırakıyoruz ve gereksiz boşlukları sil ve 1 tane boşluk bırak diyoruz
print("Temzilenmiş metin:", metin)
```

```python
# Metni küçük harfe dönüştürmek
metin= metin.lower()
print("Temzilenmiş Metin", metin)
```

```python
# Özel karakterleri kaldırmak
metin = re.sub(r"[^a-zA-ZçğıöşüÇĞİÖŞÜ0-9\s]", "", metin)
print("Temizlenmiş Metin: ", metin)
```

```python

from spellchecker import SpellChecker

spell=SpellChecker(language="en")
text="Ths is an axample of the txt with sme speling errors"

corrected_words=[]
for word in text.split():
  corrected_words.append(spell.correction(word))
corrected_text=" ".join(corrected_words)
print("Düzeltilmiş Metin:", corrected_text)
```
