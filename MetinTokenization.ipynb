• NLTK (Natural Language Toolkit), Python programlama
dilinde doğal dil işleme (NLP) görevlerini gerçekleştirmek için
yaygın olarak kullanılan açık kaynaklı bir kütüphanedir. Dil
işleme, metin analizi, ve metin madenciliği gibi çok çeşitli NLP
görevlerinde güçlü araçlar sunar.
** Özellikler:**
*   Tokenization: Metinleri kelime veya cümleye böler.
*   Part-of-Speech (POS) Etiketleme: Kelimelerin gramer türlerini
belirler.
*   Özellik Çıkartma: Metinlerden anlamlı özellikler çıkarır.
*   Makine Öğrenmesi: NLP ve dil modelleme için araçlar sağlar.

```python
import nltk
nltk.download ("punkt")
#nltk.download("punkt_tab")

from nltk.tokenize import word_tokenize

text="Merhaba dünya! Bugün Jimin nereye gitti?"
tokens= word_tokenize(text) #metni kelime kelime parçalara ayırıyoruz
print(tokens)
```

```python
from nltk.tokenize import sent_tokenize  #cümle bazlı parçalara ayırır

text="Merhaba Dünya! Nasılsın? Ben de iyiyim, teşekkürler."
sentences =sent_tokenize(text)
print(sentences)
```

```python
import spacy #en destek verir
nlp= spacy.load("en_core_web_sm")

text ="Apple INC. is planning to open a new store in the US"
doc=nlp(text)

tokens=[token.text for token in doc]  #kelime bazlı ayrım
print(tokens)
```

```python
import spacy #daha hızlı işlem yapmamızı sağlıyor nltk'ye göre
nlp= spacy.load("en_core_web_sm")

text="Artificial intelligence is amazing! It is transforming many industries"

doc= nlp(text)

sentences=[sent.text for sent in doc.sents] # cümle bazlı parça ayrım

print(sentences)
```
