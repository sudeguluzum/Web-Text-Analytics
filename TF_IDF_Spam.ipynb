```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#veri seti yükleme
df= pd.read_csv("sms_spam.csv")
df.info()
df.head()
```

```python
documents = df["text"]
```

```python
#metin temizleme
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def clean_text(text):
  text=text.lower() #büyük küçük harf çevirme
  text=re.sub(r"\d+", "", text) #rakamları temizle
  text= re.sub(r"[^\w\s]", "", text) #özel karekteri kaldır
  stop_words = set(stopwords.words('english')) #stop words listesini tanımla
  text = " ".join([word for word in text.split()if len(word)>2 and word not in stop_words]) #kısa keliemleri ve stop wordsleri temizle

  return text # temizlenmiş text'i return et

#metinleri temizle
cleaned_doc =[clean_text(row) for row in documents]

```

```python
#tfidf
vectorizer = TfidfVectorizer()
X= vectorizer.fit_transform(cleaned_doc)

#kelime kümesi
feature_names= vectorizer.get_feature_names_out()
tfidf_score = X.mean(axis=0).A1 #her kelimenin ortlama tf_idf değerleri

print("kelime kümesi: ", feature_names)
```

```python
#tdfid skolarııiçeren bir df oluştur

df_tfidf= pd.DataFrame({"word":feature_names, "tfidf_score": tfidf_score})

#skorları sırala ve sonuçları inceler
df_tfidf_sorted= df_tfidf.sort_values(by="tfidf_score", ascending=False)
df_tfidf_sorted.head(20)
```
