```python
import re
import string
from bs4 import BeautifulSoup

def metin_temizle(metin):
  metin=BeautifulSoup(metin,"html.parser").get_text()
  metin = re.sub(r"http\S+", "",metin,flags=re.MULTILINE)
  metin = metin.translate(str.maketrans("","",string.punctuation)) # metin = re.sub(r"[^a-zA-ZçğıöşüÇĞİÖŞÜ0-9\s]", "", metin)
  metin = metin.lower()
  metin = " ".join(metin.split())
  return metin

ornek_metin="""
<div> Bugun havva cok guzel !!!
Ziyaret et: https://wwwwhava.com :)
Bu #yaz $100'a mal oldu...
</div>
"""

temizlenmis_metin= metin_temizle(ornek_metin)
print("Temizlenmiş Veri:", temizlenmis_metin )
```
