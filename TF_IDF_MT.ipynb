**TF-IDF** özellik çıkarım(feature extraction) yöntemidir. Belgedeki terimin önem derecesini, hem belgedeki sıklığını hem de tüm koleksiyondaki ayırt ediciliğini dikkate alarak hesaplar


*   Term Frequency() bir kelimenin bir belgede ne kadar sık geçtiğini ölçer.

*   Invserse Document Frequency(bir kelimenin tüm belgedeki yaygınlığını ölçer. bir kelime çok belgede geçiyorsa, o kelime çok fazla bilgi sağlamaz)



```python
import pandas as pd
import numpy as np
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#örnek belge oluşturma
documents=["Makine öğrenmesi verilerden desenler çıkarır.",
           "Derin öğrenme makine öğrenmesinin bir alt dalıdır.",
           "Doğal dil işleme metinleri analiz eder.",
           "Makine öğrenmesi ve yapay zeka günümüzde çok öenmlidir."]
```

```python
#vektorizer tanımlama
tdidf_vectorizer=TfidfVectorizer()

#metinleri sayısal hale çevirme
X= tdidf_vectorizer.fit_transform(documents)

#kelime kümesi
feature_names= tdidf_vectorizer.get_feature_names_out()

print("kelime kümesi: ", feature_names)
```

```python
#vektör temsili
vektor_temsili = X.toarray()
print(f"tf-idf: {vektor_temsili}")

df_tfidf= pd.DataFrame(vektor_temsili, columns=feature_names)

df_tfidf.head()
```
